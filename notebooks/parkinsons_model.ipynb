```python
#importing  required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
#loading dataset

df = pd.read_csv('../datasets/parkinsons.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
df = df.drop('name', axis=1)
```

```python
df['status'].value_counts()
```

```python
#corelation heatmap

plt.figure(figsize=(12,10))
sns.heatmap(df.corr(),cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#split features and target

X = df.drop('status',axis=1)
y = df['status']
```

```python
# train test split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
#feature scaling

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# logistic Regression

lr = LogisticRegression(max_iter=2000)
lr.fit(X_train_scaled,y_train)

y_pred_lr = lr.predict(X_test_scaled)

print("Logistic Regression Accuracy:",accuracy_score(y_test,y_pred_lr))
print(classification_report(y_test,y_pred_lr))
```

```python
# Random forest

rf = RandomForestClassifier(n_estimators=300,random_state=42)
rf.fit(X_train,y_train)

y_pred_rf = rf.predict(X_test)

print("Random Forest Accuracy:",accuracy_score(y_test,y_pred_rf))
print(classification_report(y_test,y_pred_rf))
```

```python
#SVM

svm = SVC(kernel='rbf',probability=True)
svm.fit(X_train_scaled,y_train)

y_pred_svm = svm.predict(X_test_scaled)

print("SVM Accuracy:",accuracy_score(y_test,y_pred_svm))
print(classification_report(y_test,y_pred_svm))
```

```python
# probability prediction

sample = X_test.iloc[0:1]

prob = svm.predict_proba(scaler.transform(sample))[0][1]
pred = svm.predict(scaler.transform(sample))[0]

print("Prediction:",pred)
print("Probability of Parkinson's:",prob)

```

```python
def get_risk_level(prob):
    if prob < 0.3:
        return "Low"
    elif prob < 0.7:
        return "Medium"
    else:
        return "High"

print("Risk Level:", get_risk_level(prob))
```

```python
import pickle

with open('../models/parkinsons_model.pkl', 'wb') as f:
    pickle.dump(svm, f)

with open('../models/parkinsons_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

with open('../models/parkinsons_columns.pkl', 'wb') as f:
    pickle.dump(X.columns.tolist(), f)

print("Parkinson's model, scaler, and columns saved!")
```
